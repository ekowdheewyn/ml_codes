```python
import matplotlib.pyplot as plt
from keras import preprocessing
from keras.models import load_model
import numpy as np
%matplotlib inline
```

```python
model = load_model('malariamodel2.h5')
```

```python
class_names = ['Infected', 'Uninfected']
```

```python
width = 96
height = 96
```

```python
print("Input file name to predict: ")
file_name = str(input())
```

```python
type_1 = preprocessing.image.load_img('./data/'+file_name+'.png',
                                      target_size=(width, height))
plt.imshow(type_1)
plt.show()

type_1_X = np.expand_dims(type_1, axis=0)

predictions = model.predict(type_1_X)

print('The type predicted is: {}'.format(class_names[np.argmax(predictions)]))
```
